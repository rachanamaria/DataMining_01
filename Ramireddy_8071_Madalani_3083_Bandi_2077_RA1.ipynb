# Task 3: Visualization (30 points)

```r
# Task 3-a Display a plot where educated applicants are granted loans
cat("Task 3-a:\n")
sns_pal <- RColorBrewer::brewer.pal(9, "Blues")[3:9]
#bar<-subset(df, Education == "Graduate" & Loan_Status == "Y")
bar<-subset(df, Loan_Status == "Y")
library(ggplot2)
ggplot(bar, aes(x = Education, fill = Loan_Status)) + 
  geom_bar(position = "dodge") +
  scale_fill_manual(values = sns_pal) +
  labs(x = "Education", y = "Loan Approval", fill = "Loan Approved") +
  ggtitle("Loan Approval by Education")
```

```r
# Task 3-b Create a pie chart for Property_area and display percentages in legend respectively
cat("Task 3-b:\n")
library(ggplot2)
library(scales)
sns_pal <- RColorBrewer::brewer.pal(9, "Blues")[3:9]
prop_area <- round(prop.table(table(df$Property_Area)) * 100, 2)
pie(prop_area, main = "Property Area", col = sns_pal)
legend("topleft", fill = sns_pal, 
       cex = 0.9,title = "Property Area", 
       legend = paste0(names(prop_area),":",c(prop_area), "%"))
```

```r
# Code and explaination for Task4
library(scales)
cat("Task 4-a:\n")
blues <- RColorBrewer::brewer.pal(4, "Blues")[2:4]
ggplot(df, aes(x = Gender, fill = Loan_Status)) +
  geom_bar() +
  labs(title = "Loan Approval Rates by Gender", x = "Gender", y = "Number of Loans") +
  scale_fill_manual(values = blues, name = "Loan Status",
                    labels = c("Approved", "Denied")) +
  theme_minimal()
cat("\nFrom the graph, Among those who received loan approval, the approval rate for male applicants was greater than\n")
cat("the approval rate for female applicants, but the denial rate was higher for female applicants.\n")
cat("This shows that gender biases may exist in the loan acceptance process.")

df$Income_Group <- cut(df$ApplicantIncome, breaks = c(0, 2500, 5000, 7500, 10000, Inf),
                         labels = c("0-2500", "2501-5000", "5001-7500", "7501-10000", ">10000"))
ggplot(df, aes(x = "", y = LoanAmount, fill = Income_Group)) +
  geom_boxplot() +
  labs(x = "Income", y = "Loan Amount", fill = "Income Group") +
  ggtitle("Relationship between Income and Loan Amount") +
  scale_fill_brewer(palette = "Blues")
cat("\nTask 4-b:\n")
cat("\nHigher earners have a better chance of getting bigger loans approved.\n")
cat("Yet, there is also a significant degree of difference in the loan amounts\n")
cat("authorized for people with comparable salaries, implying that loan acceptance\n")
cat("decisions may include other considerations besides income.")
```
